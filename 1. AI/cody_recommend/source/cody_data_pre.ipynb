```python
#9조 뭐입지니 모델링 팀원 김형우 produce
#dataPre.py: 라벨링된 데이터 파일을 가지고
#학습에 필요한 데이터 파일을 생성한다.

import pandas as pd
import numpy as np
import csv

#feature들의 값을 지정해주기 위한 리스트, index+1 값으로 mapping 진행
cate = ["long_blouse/long_chino","short_tee/short_skirt","long_sports/sweats","short_sports/leggings",
       "short_shirt/long_chino","long_shirt/long_chino","short_blouse/long_chino","short_tee/short_chino",
       "long_sports/leggings","short_sports/sweats","long_tee/long_denim","long_tee/long_skirt",
       "short_tee/long_denim","short_tee/long_skirt",
        "short_tee/long_chino","long_tee/long_chino","short_tee/short_denim"]#추가 부분
weather = ["sunny","cloudy","rainy"]
style = ["business","casual","sports"]
sex = ["male","female"]
boShape = ["thin","normal","fat"]

#라벨링된 파일을 열고 읽어온다.
f = open('cody_raw.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
#변경된 내용을 저장할 리스트
lines=[]

#all lines 진행
for line in rdr:
    if(line[0] == "temp"): #저장된 텍스트 값을 index+1 값으로 지정
        lines.append(line)
    else:
        line[1] = weather.index(line[1])+1
        line[3] = sex.index(line[3])+1
        line[6] = boShape.index(line[6])+1
        line[7] = style.index(line[7])+1
        line[8] = cate.index(line[8])+1
        lines.append(line)
#변경된 내용을 저장한다.
f = open('data_conv.csv','w',newline='')
wr = csv.writer(f)
wr.writerows(lines)
f.close()
```

```python
#변환된 값의 파일을 읽어온다.
f = open('data_conv.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
#변경된 내용을 저장할 리스트
lines=[]

#all lines 진행, y값 col 삭제
for line in rdr:
    del line[8]
    lines.append(line)

#변경된 내용을 저장한다.
f = open('data_x.csv.csv','w',newline='')
wr = csv.writer(f)
wr.writerows(lines)
f.close()
```
